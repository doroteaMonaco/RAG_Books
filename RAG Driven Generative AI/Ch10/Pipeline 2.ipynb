```python
import requests

def download(directory, filename):
    base_url = 'https://raw.githubusercontent.com/Denis2054/RAG-Driven-Generative-AI/main/'
    file_url = f"{base_url}{directory}/{filename}"
    print("Trying to download:", file_url)
    try:
        response = requests.get(file_url)
        response.raise_for_status()
        with open(filename, 'wb') as f:
            f.write(response.content)
        print(f"Downloaded '{filename}' successfully.")
    except Exception as e:
        print(f"Failed to download '{filename}': {e}")
```

```python
import time 
from sentence_transformers import SentenceTransformer

embedding_model = "all-MiniLM-L6-v2"


def create_embeddings(chunks, embedding_model):
    model = SentenceTransformer(embedding_model)
    start_time = time.time()
    embeddings = model.encode(chunks, show_progress_bar=True)
    end_time = time.time()
    return embeddings.tolist()
```

```python
# Optimized batch embedding processing
print(f"Embedding {len(chunks)} chunks...")
embeddings = create_embeddings(chunks, embedding_model)
pure_embeddings = list(embeddings) # Ensure pure_embeddings is always in sync
print(f"All chunks processed. Total embeddings successfully created: {len(pure_embeddings)}.")
```

```python
import time
import pinecone
if index_name in pc.list_indexes().names():
    idx_info = pc.describe_index(index_name)
    if idx_info.dimension != 384:
        print(f"Deleting index {index_name} because dimension {idx_info.dimension} != 384")
        pc.delete_index(index_name)
        time.sleep(2)

if index_name not in pc.list_indexes().names():
    print(f"Creating index {index_name} with dimension 384")
    pc.create_index(
        index_name,
        dimension=384, 
        metric='cosine',
        spec=spec
    )
    time.sleep(1)
```

```python
start_time = time.time()

min_len = min(len(df), len(pure_embeddings))
if len(df) != len(pure_embeddings):
    print(f"WARNING: Length mismatch! df: {len(df)}, embeddings: {len(pure_embeddings)}. Using minimum length {min_len}.")

data_for_upsert = [
    {
        "id": str(df['ID'].iloc[i]),  # Use existing IDs as string
        "values": pure_embeddings[i],
        "metadata": {
            "text": df['Comment'].iloc[i],
            "frame_number": df['FrameNumber'].iloc[i],
            "file_name": df['FileName'].iloc[i]
        }
    }
    for i in range(min_len)
]

batch_size = 100  
def upsert_in_batches(data, batch_size):
    for start in range(0, len(data), batch_size):
        end = start + batch_size
        batch = data[start:end]
        try:
            index.upsert(vectors=batch)
            print(f"Upserted batch from index {start} to {end - 1}")
        except Exception as e:
            print(f"Failed to upsert batch from index {start} to {end - 1}: {e}")

upsert_in_batches(data_for_upsert, batch_size)
response_time = time.time() - start_time
print(f"Total upsertion time: {response_time:.2f} seconds")
```

### Querying vector store after upserting the dataset

```python
k = 1
```

```python
query_text = "Find a basketball player that is scoring with a dunk."
```

```python
import time

# Start timing before the request
start_time = time.time()

# Target vector
query_embedding = create_embeddings([query_text], embedding_model)[0]  # Get the first (and only) embedding from the list

query_results = index.query(vector=query_embedding, top_k=k, include_metadata=True)  # Request metadata

print("Query Results:")
for match in query_results['matches']:
    print(f"ID: {match['id']}, Score: {match['score']}")

    if 'metadata' in match:
        metadata = match['metadata']
        text = metadata.get('text', "No text metadata available.")
        frame_number = metadata.get('frame_number', "No frame number available.")
        file_name = metadata.get('file_name', "No file name available.")

        print(f"Text: {text}")
        print(f"Frame Number: {frame_number}")
        print(f"File Name: {file_name}")
    else:
        print("No metadata available.")

response_time = time.time() - start_time
print(f"Querying response time: {response_time:.2f} seconds")  
```

```python
def display_video(file_name):
  with open(file_name, 'rb') as file:
      video_data = file.read()

  video_url = b64encode(video_data).decode()

  html = f'''
  <video width="640" height="480" controls>
    <source src="data:video/mp4;base64,{video_url}" type="video/mp4">
  Your browser does not support the video tag.
  </video>
  '''
  HTML(html)
  return HTML(html)
```
